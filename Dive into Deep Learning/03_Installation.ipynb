# 安装

本地安装（现在是CPU安装，等过一段时间才用到GPU）

【可选】  使用 conda / miniconda 环境

  conda env remove d2l-zh

  conda create -n -y d2l-zh python=3.8 pip  #(此处用3.8python)

  conda activate d2l-zh

安装需要的包

  pip install jupyter d2l torch torchvision

下载代码并执行 （这部分是Linux指令）

  wget http://zh-v2.d2l.ai/d2l-zh.zip

  unzip d2l-zh.zip

  jupyter notebook
