# 06 矩阵计算

**标量导数**

+ 导数是切线的斜率

该部分即曾经学过的函数求导及其运算规则

**亚导数**

+ 将导数扩展到不可微的函数

如：  y=|x|在x=0位置的导数，在x=0的位置其导数不唯一，可以在-1到1之间取任意的值

**梯度**

+ 将导数扩展到向量

1. x是标量，y是标量，求导为标量
2. x是标量，y是向量，求导为向量
3. x是向量，y是标量，求导为向量
4. x是向量，y是向量，求导为矩阵

这个被称之为分子布局符号，反过来的版本叫分母布局符号

行向量关于标量的导数是一个行向量，列向量关于标量的导数也是列向量；当x为向量时导数为行向量，当y为向量时导数为列向量

当x和y都为向量时，其导数就变成了矩阵形式
