```python
import torch
# 导入pytorch
```

```python
# 张量表示一个数值组成的数组，该数组可能有多个维度
x = torch.arange(12) # 用arange创造一个 行向量x
x
```

```python
# 运算符
```

```python
# 常见的标准算数运算符都可以杯升级为 按元素运算
x = torch.tensor([1.0,2,4,8])
y = torch.tensor([2,2,2,2])
x+y , x-y , x*y , x/y , x**y # **为求幂
```

```python
# 当然也就能按元素方式应用更多的计算
torch.exp(x)
#除了按元素计算以外，我们也能执行线性代数运算，在后面会讲到
```

```python
# 我们也可以把多个张量连结在一起，根据dim值不同（维度不同），有不同连结方式：
#【示例中】按行连结（轴-0，即形状的第一个元素）；按列连结（轴-1，即形状的第二个元素）
x = torch.arange(12,dtype = torch.float32).reshape((3,4))
y = torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
torch.cat((x,y),dim = 0) , torch.cat((x,y),dim = 1) #按行连结 - 3（连接后为3+3行） 和 按列连结 - 4（连结后为4+4列）
```

```python
# 通过逻辑运算符来构建二元张量 
# 【示例 x==y】在对应位置上，若x==y则为True，反之为False
x == y
```

```python
# 对张量中的所有元素进行求和会产生一个只有一个元素的张量
x.sum()
```

```python
# 广播机制
```

```python
# 即使形状不同，我们仍然可以通过广播机制来执行按元素操作，执行方式如下：
# 1.通过适当复制元素来扩展一个或两个数组，以便在转换之后，两个张量具有相同的形状
# 2.对生成的数组执行按元素操作
a = torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
a,b
```

```python
a + b
# 广播机制 先将两个向量复制成(3,2)后再进行运算
```

```python
# 索引和切片
```

```python
#前者为索引，可以用-1选择最后一个元素，当然-2就是倒数第二个元素
#后者为切片，x[a,b]表示选择[a,b)区间内的元素
x[-1],x[1:3]
```

```python
#除了读取以外，我们还可以通过指定索引来将元素写入矩阵
x[1,2] = 9
x
```

```python
#如果要给多个元素赋同一值，我们只需要索引所有的元素，然后为它赋值即可
x[0:2,:] = 12
x
```

```python
# 节省新内存
```

```python
# 运行一些操作可能会导致为新结果分配内存
before = id(y)
y = y + x
id(y) == before
# id() 函数用于返回对象的唯一标识符，通常词标识符就是该对象在内存中的 地址 
```

```python
# 也可以执行原地操作——使用切片表示法将操作的结果分配给先前分配的数组
z = torch.zeros_like(y)
print('id(z):',id(z))
z[:] = z + y
print('id(z):',id(z))
```

```python
#如果在后续计算中没有重复使用 x，我们也可以使用 x[:]=x+y 或 x+=y 减少操作的内存开销
before = id(x)
x += y
id(x) == before
```

```python
#创建一个人工数据集，并储存在csv（逗号分隔值）文件中
import os

os.makedirs(os.path.join('..', 'data'), exist_ok=True)
data_file = os.path.join('..', 'data', 'house_tiny.csv')
with open(data_file, 'w') as f:
    f.write('NumRooms,Alley,Price\n')  # 列名
    f.write('NA,Pave,127500\n')  # 每行表示一个数据样本
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
#该操作是在当前工作目录的上一级创建一个data文件夹，并在其中生成一个house_tiny.csv的小型示例数据集
```

```python
#常用pandas来读取数据
import pandas as pd

data = pd.read_csv(data_file)
print(data)
data
```

```python
# 处理缺失值
```

```python
# 为了处理缺失的数据，典型的方法包括有 插值 和 删除 。
#此处是插值示例
inputs, outputs = data.iloc[:, 0:2], data.iloc[:, 2]
inputs = inputs.fillna(inputs.mean(numeric_only=True)) #这里mean后面的括号内要加上这个条件，否则会有数据类型的报错
#fillna是按列对其填充，其括号中的内容则为填充值（.mean为计算安全均值）
print(inputs)
inputs
```

```python
#对于inputs中的类别值或离散值，我们将“NaN”视为一个类别
inputs = pd.get_dummies(inputs, dummy_na=True)
print(inputs)
inputs
```

```python
#转换为张量格式
```

```python
#现在的inputs和outputs中的所有条目都是数值类型，他们就能转换为张量格式
X = torch.tensor(inputs.to_numpy(dtype=float))
y = torch.tensor(outputs.to_numpy(dtype=float))
X, y
```
